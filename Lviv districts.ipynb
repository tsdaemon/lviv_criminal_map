```python
import http.client as http
from urllib.parse import quote
import json
from time import sleep

def wikimapia_request(function, additional_query):
    key = "B0B82BBD-9F7D7E6E-FEFF457C-41A336F9-9C94B3C8-1EC8D213-E1D3CF06-38539DF1"

    query = "/?function={0}&key={1}&format=json&".format(function, key) + additional_query

    conn = http.HTTPConnection("api.wikimapia.org")
    conn.request("GET", query)
    res = conn.getresponse()
    res_str = res.read().decode("utf-8")
    return json.loads(res_str)
```

```python
import codecs
categories = {
    'neighbourhood': 45057,
    'historic': 46460,
    'district': 45535
}
lat1 = 49.5
lat2 = 50.5
lng1 = 23.5
lng2 = 24.5

def dump_polygons(category_name, category_id):
    result = {[]}
    left = 100
    page = 1
    while(left > 0):
        function = "box"
        additional_query="lon_min={0}&lat_min={1}&lon_max={2}&lat_max={3}&category={4}&count=100&page={5}".format(lng1, lat1, lng2, lat2, category_id, page)
        response = wikimapia_request(function, additional_query)
        result["polygons"] += response["folder"]
        left = int(response["found"]) - int(response["count"])*page
        page += 1
        
    with codecs.open("./polygons/{0}.json".format(category_name), 'w', 'utf8') as f:
        json.dump(result, f, indent = 2, ensure_ascii = False)

for cat_name in categories.keys():
    cat_id = categories[cat_name]
    dump_polygons(cat_name, cat_id)
```

```python
def inside_polygon(x, y, points):
    """
    Return True if a coordinate (x, y) is inside a polygon defined by
    a list of verticies [(x1, y1), (x2, x2), ... , (xN, yN)].

    Reference: http://www.ariel.com.au/a/python-point-int-poly.html
    """
    n = len(points)
    inside = False
    [p1x, p1y] = points[0]
    for i in range(1, n + 1):
        [p2x, p2y] = points[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xinters = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside
```
